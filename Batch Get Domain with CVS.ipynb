## Library Imports

```python
import requests
import json
import pandas as pd
```

## Importing Our List of App Market ID

```python
stocks = pd.read_csv('Vungle ad_transactions 2021-10-05T1747-android - 副本.csv')
```

```python
marketID=stocks['marketid']
```

## Adding columns to a Pandas DataFrame

```python
my_columns = ['marketid', 'AppAdsUrl']
```

##  Looping Through The MarketID in Our List of CSV
1. If dataJson[1] is not work, try dataJson[0]
2. Change platform iOS/Android before run

```python
final_dataframe = pd.DataFrame(columns = my_columns)

for marketid in stocks['marketid']:
    api_url = f'http://api.adtimizer.co/ios/app-ads-txt?id={marketid}'
    data = requests.get(api_url)
    if data.status_code == 200:
        dataJson=data.json()
        final_dataframe = final_dataframe.append(
                            pd.Series([
                                        marketid,
                                        dataJson[0]['AppAdsUrl'],],
                                        index = my_columns), 
                                        ignore_index = True)

    elif data.status_code == 204:
       
        final_dataframe = final_dataframe.append(
                            pd.Series([
                                        marketid,
                                        data.status_code,],
                                        index = my_columns), 
                                        ignore_index = True)
    
final_dataframe
```
